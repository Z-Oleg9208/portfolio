# Восстановление золота из руды

Подготовка прототипа модели машинного обучения для «Цифры». Компания разрабатывает решения для эффективной работы промышленных предприятий.

Модель должна предсказать коэффициент восстановления золота из золотосодержащей руды. Использя данные с параметрами добычи и очистки.

Модель поможет оптимизировать производство, чтобы не запускать предприятие с убыточными характеристиками.

Описание данных

**Технологический процесс**
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    - Xanthate — ксантогенат (промотер, или активатор флотации);
    - Sulphate — сульфат (на данном производстве сульфид натрия);
    - Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

**Параметры этапов**
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

**Наименование признаков**

Наименование признаков должно быть такое:

[этап].[тип_параметра].[название_параметра]

Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:

- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики

Возможные значения для блока [тип_параметра]:

- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака `rougher.output.recovery`. Найдем MAE между нашими расчётами и значением признака.

```python
C = train['rougher.output.concentrate_au']
F = train['rougher.input.feed_au']
T = train['rougher.output.tail_au']
```

```python
recovery_tabl = train['rougher.output.recovery'].to_frame()
recovery_tabl['recovery'] = (C * (F - T))/(F * (C - T)) * 100
mae = abs(sum(recovery_tabl['rougher.output.recovery'])
                                   - sum(recovery_tabl['recovery'])) / len(recovery_tabl['recovery'])
print(f'MAE = {mae:.5f}')
```

#### Вывод 

Средняя абсолютная ошибка (MAE) между `rougher.output.recovery` и расчетным значением recovery равняется 0 с точностью до пяти знаков после запятой. Это означает, что `rougher.output.recovery` расчитано верно. **Дальнейшее использование данного целевого признака для обучения модели является корректным.**

## Модель

```python
# функция для вычисления итоговой sMAPE.

def sMAPE(data, predicts):
    df_sum = ((abs(data - predicts))/((abs(data) + abs(predicts))/2)).sum()
    smape = ((abs(data - predicts))/((abs(data) + abs(predicts))/2)).sum()/len(data) *100
    return smape

def scorrer():
    return make_scorer(sMAPE, greater_is_better=False)
```

### Подберем модель для предсказания эффективности обогащения чернового концентрата (rougher.output.recovery)

```python
pipelin = Pipeline([('scaler', StandardScaler()),
                    ('model', DecisionTreeRegressor())])

params_1 = [{
    'model': [DecisionTreeRegressor(random_state=42)],
    'model__max_depth':range(1,10),
},
    {
        'scaler': [StandardScaler(), MinMaxScaler()],
        'model': [RandomForestRegressor(random_state=42)],
        'model__max_depth':range(1,5),
        'model__n_estimators':range(1,50,5),
    }]

model_rougherv = GridSearchCV(pipelin,
                       params_1,
                       cv=5,
                       scoring=scorrer(),
                       n_jobs=-1
                      )

model_rougherv.fit(features_rougherv_train, target_rougherv_train)

model_rougherv.best_score_
```

```python
params_2 = [{
       'scaler': [StandardScaler(), MinMaxScaler()],
       'model': [LinearRegression()], 
    },
    
    {
      'scaler': [StandardScaler(), MinMaxScaler()],
      'model': [Lasso()],
      'model__alpha': [n/10 for n in range(1, 20)],
      'model__max_iter': range(1000, 10000, 10),
      'model__tol': (1e-6, 1e+6, 'log-uniform')
    }
]

model_rougherv_lin = RandomizedSearchCV(pipelin,
                       params_2,
                       cv=7,
                       scoring=scorrer(),
                       n_iter=50,
                       n_jobs=-1
                      )

model_rougherv_lin.fit(features_rougherv_train, target_rougherv_train)

model_rougherv_lin.best_score_
```
